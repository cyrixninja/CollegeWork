```python
# Data Visualization in python using matplotlib
```

```python
# 8.1 Create DataFrame with 100 records, 2 columns and random numbers. Use this data to create a line plot.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(np.random.randn(100, 2), columns=['A', 'B'])
df.plot()
plt.show()
```

```python
# 8.2  Create DataFrame with 100 records, 2 columns and random numbers. Use this data to create a scatter plot.

df = pd.DataFrame(np.random.randn(100, 2), columns=['A', 'B'])
df.plot(kind='scatter', x='A', y='B')
plt.show()
```

```python
# 8.3  Create a numpy array with 100 random numbers and plot a histogram chart on it.

data = np.random.randn(100)
plt.hist(data, bins=10)
plt.show()
```

```python
# 8.4 Create sample data and plot pie chart using it.

data = [5, 10, 15, 20, 25]
labels = ['A', 'B', 'C', 'D', 'E']
plt.pie(data, labels=labels)
plt.show()
```
