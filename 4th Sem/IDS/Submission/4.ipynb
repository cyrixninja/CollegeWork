```python
# Set-4 Data Manipulation with Pandas

# 4.1 Write a program to add, subtract, multiply and divide two panda series
import pandas as pd
import numpy as np

def add_sub_mul_div():
    # Create two panda series
    series1 = pd.Series([2, 4, 6, 8, 10])
    series2 = pd.Series([1, 3, 5, 7, 9])

    # Add two panda series
    add = series1 + series2
    print("Addition of two panda series:")
    print(add)

    # Subtract two panda series
    sub = series1 - series2
    print("\nSubtraction of two panda series:")
    print(sub)

    # Multiply two panda series
    mul = series1 * series2
    print("\nMultiplication of two panda series:")
    print(mul)

    # Divide two panda series
    div = series1 / series2
    print("\nDivision of two panda series:")
    print(div)

add_sub_mul_div()

```

```python
# 4.2 Perform the following operations for DataFrame:
# a. Create a dataframe for the following data:
# Input:
# exam_data = {
# 'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily',
# 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
# 'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
# 'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
# 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes'] }
# labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
# b. Select and display the 'name' and 'score' columns
# c. Select 'score' and ‘qualify’ columns in rows 1, 3, 5 and 6
# d. Display the rows where the number of attempts > 2
# e. Display the rows where the score is missing, i.e. is NaN
# f. Display rows where the score is between 15 and 20 (inclusive).
# g. Change the score in row 'd' to 11.5.
# h. Create a dataframe grouped by attempts and display the mean of score for each group.

exam_data = {   'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
                'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
                'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
                'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes'] }
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

def data_frame_operations():

    # Create a dataframe
    df = pd.DataFrame(exam_data, index=labels)
    print("\nDataframe:")
    print(df)

    # Select and display the 'name' and 'score' columns
    print("\nName and Score columns:")
    print(df[['name', 'score']])

    # Select 'score' and ‘qualify’ columns in rows 1, 3, 5 and 6
    print("\nScore and Qualify columns in rows 1, 3, 5 and 6:")
    print(df.loc[df.index[[1, 3, 5, 6]], ['score', 'qualify']])

    # Display the rows where the number of attempts > 2
    print("\nRows where the number of attempts > 2:")
    print(df[df['attempts'] > 2])

    # Display the rows where the score is missing, i.e. is NaN
    print("\nRows where the score is missing:")
    print(df[df['score'].isnull()])

    # Display rows where the score is between 15 and 20 (inclusive)
    print("\nRows where the score is between 15 and 20:")
    print(df[df['score'].between(15, 20)])

    # Change the score in row 'd' to 11.5
    print("\nChange the score in row 'd' to 11.5:")
    df.loc['d', 'score'] = 11.5
    print(df)

    # Create a dataframe grouped by attempts and display the mean of score for each group
    print("\nDataframe grouped by attempts and display the mean of score for each group:")
    print(df.groupby('attempts')['score'].mean())

data_frame_operations()
```

```python
# 4.3 Read any dataset into dataframe and display following information:
# a. Number of rows
# b. Number of attributes
# c. Properties of each attributes
# d. Number of missing values for each attributes
# e. Total number of missing values
# f. Replacing missing value with following:
#    i. constant value
#   ii. mean value
#   iii. fill using backfill
# g. Drop null values

df = pd.read_csv('missing-employees.csv')
print("\nNumber of rows:")
print(len(df))

# b.
print("\nNumber of attributes:")
print(len(df.columns))

# c.
print("\nProperties of each attributes:")
print(df.dtypes)

# d.
print("\nNumber of missing values for each attributes:")
print(df.isnull().sum())

# e.
print("\nTotal number of missing values:")
print(df.isnull().sum().sum())

# f.
# i. constant value
print("\nReplacing missing value with constant value:")
print(df.fillna(0))

# ii. mean value
print("\nReplacing missing value with mean value:")
numeric_columns = df.select_dtypes(include=[np.number]).columns
print(df[numeric_columns].fillna(df[numeric_columns].mean()))

# iii. fill using backfill
print("\nReplacing missing value with backfill:")
print(df.fillna(method='bfill'))

# g.
print("\nDrop null values:")
print(df.dropna())
```
